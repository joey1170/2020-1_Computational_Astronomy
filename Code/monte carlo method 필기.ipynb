```python
import numpy as np
import matplotlib.pyplot as plt

def randwalk(N):                       
    x = np.random.rand(N)
    R = np.count_nonzero(x >= 0.5)
    position = 2*R - N
    return position

def normal(x, D, t):
    p = np.exp(-(x**2)/(4*D*t))/np.sqrt(4*np.pi*D*t)
    return p

l = 1
tau = 1
D = 0.5*(l**2)/tau
n = 10**3
N = 10**4
N2 = 10**5
t = N*tau
t2 = N2*tau

res = np.zeros(n)
for i in range(n):
    res[i] = randwalk(N)

res2 = np.zeros(n)
for i in range(n):
    res2[i] = randwalk(N2)    
    
    
mean = np.mean(res)
var = np.var(res)
print('blue one : mean is {}, var/num is {}'.format(mean, var/N))

mean2 = np.mean(res2)
var2 = np.var(res2)
print('red one : mean is {}, var/num is {}'.format(mean2, var2/N2))

range1 = 1000



x = np.linspace(-range1, range1, 1000)
pdf = normal(x, D, t)
pdf2 = normal(x, D, t2)

bins = np.linspace(-range1, range1, 100)
plt.hist(res, bins, density = 1, color = 'blue', alpha = 0.5, label = '10000 steps')
plt.hist(res2, bins, density = 1, color = 'red', alpha = 0.5, label = '100000 steps')
plt.xlim(-1000,1000)
plt.legend()
plt.plot(x, pdf)
plt.plot(x, pdf2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x)

def MonteIntegrate(a, b, N, f):
    x = (b - a)*np.random.rand(N)
    I = (b - a)*np.sum(f(x))/N
    
    return I

a = 0
b = 1

truevalue = np.exp(1) - 1

for i in range(2, 7):
    N = 10**i
    
    integrand = MonteIntegrate(a, b, N, f)
    error = abs(integrand - truevalue)
    print("N = 10^{}: I = {:.8f}, error = {:.8f}".format(i, integrand, error))
```

```python
#Box-Muller Method (Gaussian Deviate)

import numpy as np
import matplotlib.pyplot as plt

def BoxMuller(N, sig):               
    x1 = np.random.rand(N)
    x2 = np.random.rand(N)
    
    theta = 2*np.pi*x1
    R = np.sqrt(-2*np.log(x2))
    
    X = sig*R*np.cos(theta)
    Y = sig*R*np.sin(theta)
    
    return X, Y                      

def Gauss(x, mu, sig):
    p = np.exp(-(x - mu)**2/(2*(sig**2))) / (sig*np.sqrt(2*np.pi))
    
    return p

N = 5000
sig = 1
mu = 0

X, Y2 = BoxMuller(N, sig)
Y = np.random.normal(loc = mu, scale = sig, size = N)  

bins = np.linspace(-5, 5, 100)
plt.hist(X, bins, density = 1, color = 'red', alpha = 0.5)
plt.title('my.gasdev')
plt.plot(bins, Gauss(bins, mu, sig))


plt.show()
plt.hist(Y, bins, density = 1, color = 'blue', alpha = 0.5)
plt.plot(bins, Gauss(bins, mu, sig))
plt.title('random.normal')
plt.xlim(-5, 5)
plt.show()
```
