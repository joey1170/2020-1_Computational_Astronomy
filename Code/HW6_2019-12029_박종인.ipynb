```python

N= 5
# qdata = np.zeros((N,N))
omegadata = np.zeros((N,N))
Adata = np.zeros((N,N))
Edata = np.zeros((N,N))
lamdata = np.zeros((N,N))

wdata = [1,3,5,5,1]
# ======================= A =====
for i in range(N):
    for j in range(N):
        if i<=j:
            Adata[i,j]=0
Adata[1,0],Adata[2,0],Adata[3,0],Adata[4,0],Adata[2,1]=2.7*10**-5,3.1*10**-11, 1.7*10**-6,0,9.7*10**-5
Adata[3,1],Adata[4,1],Adata[3,2],Adata[4,2],Adata[4,3]=6.8*10**-3, 2.3*10**-1, 2.0*10**-2, 6.1*10**-4, 1.6
#=========================omegadata=====

omegadata[1,0],omegadata[2,0],omegadata[3,0],omegadata[4,0],omegadata[2,1]=0.55, 0.27, 0.254, 0.032, 1.29
omegadata[3,1],omegadata[4,1],omegadata[3,2],omegadata[4,2],omegadata[4,3]=0.763, 0.097, 1.272, 0.161, 0.58
for i in range(N):
    for j in range(N):
        if i<=j:
            omegadata[i,j]=omegadata[j,i]
            

#====================Edata============
for i in range(N):
    for j in range(N):
        if i>=j:
            Edata[i,j]=0
Edata[0,1],Edata[0,2], Edata[0,3], Edata[0,4] = 0.014, 0.038, 2.514, 5.3
Edata[1,2], Edata[1,3], Edata[1,4] = 0.024, 2.5, 5.314
Edata[2,3], Edata[2,4], Edata[3,4] = 2.476, 5.318, 2.842

#============lamdata # 단위를 위해 갑을 추후 수정 해야함 
for i in range(N):
    for j in range(N):
        if i>=j:
            lamdata[i,j]=0
lamdata[0,1], lamdata[0,2], lamdata[0,3], lamdata[0,4] = 88.256*1e-6, 32.661*1e-6, 4931.1*1e-10,0
lamdata[1,2], lamdata[1,3], lamdata[1,4] = 51.814*1e-6, 4958.9*1e-10, 2321.0*1e-10
lamdata[2,3], lamdata[2,4], lamdata[3,4] = 5006.9*1e-10, 2331.4*1e-10, 4363.2*1e-10



def calpopulation(n_e1, temp1):
    

    # ========================= qdata
    temp = temp1/10**4
    n_e = n_e1
    
    qdata = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            if j>i:
                qdata[j, i] = 8.629e-8/np.sqrt(temp)*omegadata[j, i]/(wdata[j])
                qdata[i, j] = wdata[j]/wdata[i]*qdata[j,i]*np.exp(-1.1605*Edata[i,j]/temp)


    # ====== Mdata
    
    Mdata = np.zeros((N))
    for i in range(N): 
        Mdata[i] = n_e*(sum(qdata[i])-qdata[i, i])+sum(Adata[i, :i])

    matrix = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            if j>i:
                matrix[i, j] = n_e*qdata[j, i]+Adata[j, i]
            elif i == j:
                matrix[i, j] = -Mdata[i]
            elif i == 4:
                matrix[i, j] = 1
            else:
                matrix[i, j] = n_e*qdata[j, i]
            
                
    Left = np.array([0, 0, 0, 0, 1])

    result = np.linalg.solve(matrix, Left)
    return result


print("n_e : {}cm^-3 T : {}K".format(100, 10**4))
print("n1 : {} n2 : {} n3 : {} n4 : {} n5 : {}".format(calpopulation(100, 1e4)[0],calpopulation(100, 1e4)[1],calpopulation(100, 1e4)[2],calpopulation(100, 1e4)[3],calpopulation(100, 1e4)[4]))
print("n_e : {}cm^-3 T : {}K".format(100, 2*10**4))
print("n1 : {} n2 : {} n3 : {} n4 : {} n5 : {}".format(calpopulation(100, 2*1e4)[0],calpopulation(100, 2*1e4)[1],calpopulation(100, 2*1e4)[2],calpopulation(100, 2*1e4)[3],calpopulation(100,2*1e4)[4]))
print("n_e : {}cm^-3 T : {}K".format(10, 10**4))
print("n1 : {} n2 : {} n3 : {} n4 : {} n5 : {}".format(calpopulation(10, 1e4)[0],calpopulation(10, 1e4)[1],calpopulation(10, 1e4)[2],calpopulation(10, 1e4)[3],calpopulation(10,1e4)[4]))
                             
                                                             
                                                             
```

```python
def Rcal(n4, n5):
    C1 = n4/n5
    C2 = Adata[3,2]*lamdata[2,3]+Adata[3,1]*lamdata[3,1]
    C3 = Adata[4,3]*lamdata[3,4]
    return C1*C2/C3

def calpopulation2(n_e1, temp1):
    

    # ========================= qdata
    temp = temp1/10**4
    n_e = n_e1
    
    qdata = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            if j>i:
                qdata[j, i] = 8.629e-8/np.sqrt(temp)*omegadata[j, i]/(wdata[j])
                qdata[i, j] = wdata[j]/wdata[i]*qdata[j,i]*np.exp(-1.1605*Edata[i,j]/temp)


    # ====== Mdata
    
    Mdata = np.zeros((N))
    for i in range(N): 
        Mdata[i] = n_e*(sum(qdata[i])-qdata[i, i])+sum(Adata[i, :i])

    matrix = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            if j>i:
                matrix[i, j] = n_e*qdata[j, i]+Adata[j, i]
            elif i == j:
                matrix[i, j] = -Mdata[i]
            elif i == 4:
                matrix[i, j] = 1
            else:
                matrix[i, j] = n_e*qdata[j, i]
            
                
    Left = np.array([0, 0, 0, 0, 1])

    result = np.linalg.solve(matrix, Left)
    return result[3], result[4]


n_elist = [10, 1e3, 1e5]
temperature = np.linspace(5000, 20000, 1000)

Rcal0 = np.zeros(1000)
Rcal1 = np.zeros(1000)
Rcal2 = np.zeros(1000)

for i in range(1000):
    Rcal0[i] = Rcal(calpopulation2(n_elist[0], temperature[i])[0],calpopulation2(n_elist[0], temperature[i])[1])
    Rcal1[i] = Rcal(calpopulation2(n_elist[1], temperature[i])[0],calpopulation2(n_elist[1], temperature[i])[1])    
    Rcal2[i] = Rcal(calpopulation2(n_elist[2], temperature[i])[0],calpopulation2(n_elist[2], temperature[i])[1])


plt.plot(temperature, Rcal0, label = '$n_e$ = {:.0f}'.format(n_elist[0]))
plt.plot(temperature, Rcal1, label = '$n_e$ = {:.0f}'.format(n_elist[1]))
plt.plot(temperature, Rcal2, label = '$n_e$ = {:.0f}'.format(n_elist[2]))
plt.xlabel('temperature (K)')
plt.ylabel('R')
plt.yscale('log')
plt.legend()
```

### (a) the steepest descent

```python
def func(x):
    return (x[0]**2-10*x[1])**2 + 2*x[1]**2 - 3*x[0]

def golden2D(x,dr,TOL):
    dr = dr/np.sqrt(sum(dr**2))
    R, err, loop = 0.61803399, 10., -1
    a = x - dr 
    b = x + dr
    
    while(err > TOL):
        loop += 1
        x1 = b - R*(b - a)
        x2 = a + R*(b - a)
        
        f1=func(x1)
        f2=func(x2)
        
        if(f2 > f1):
                b = x2
                
        else:
            a = x1
            
        err = sum((a - b)**2)
        err = np.sqrt(err)
        
        xmin = a
        
    return xmin, err, loop

def gradf(x):
    dfdx =4*x[0]**3-40*x[0]*x[1]-3
    dfdy = 204*x[1]-20*x[0]**2
    
    
    return np.array([dfdx, dfdy])

#==========================================================

TOL = 1.e-6
x = np.array([-5, 2])
error = 1

res = [x]

i = 0
while error > TOL:
    dr = -gradf(x)/np.sqrt(np.sum(gradf(x)**2))
    xmin, err, loop = golden2D(x, dr, TOL)
    error = np.sqrt(np.sum((xmin - x)**2))
    
    x = xmin
    res.append(x)
    
    i += 1
res = np.array(res)

#==========================================================

fig = plt.figure(1,figsize=(6,6))

x = np.linspace(-6, 5, 1000)
y = np.linspace(-5, 5, 1000)

X, Y = np.meshgrid(x, y)
E = ((X**2-10*Y)**2 + 2*Y**2 - 3*X)


dmax = np.max(E)
dmin = np.min(E)

levels = 6000*np.arange(20)/22  # (dmax-dmin)*np.arange(6000)/2900 +dmin
cmap = plt.cm.BuPu
con = plt.contourf(x, y, E, levels,cmap = cmap) 
plt.xlabel('x')
plt.ylabel('y')


plt.plot(res[:, 0], res[:, 1], color = 'blue')
plt.plot(res[:, 0], res[:, 1], 'r.')
plt.show()

print("minimum point is ({:.3f},{:.3f})".format(res[-1][0], res[-1][1]))
print("iteration {}".format(i))
```

### (b) the Powell method

```python
def func(x):
    f = (x[0]**2-10*x[1])**2 + 2*x[1]**2 - 3*x[0]
    return f

def golden2D(x,dr,TOL):
    R,err,loop=0.61803399,10.,-1
    dr = dr/np.sqrt(sum(dr**2))
    a= x-dr # initialize the range
    b= x+dr
    while(err > TOL):
        loop += 1
        x1= b-R*(b-a)
        x2= a+R*(b-a)
        f1=func(x1)
        f2=func(x2)
        if(f2>f1): b=x2
        else: a=x1
        err=sum((a-b)**2)
        err=np.sqrt(err)
    xmin=a
    return xmin,err,loop

TOL=1e-6
P=np.array([-5,2])

u1=np.array([1,0])
u2=np.array([0,1])

iter2=0
xt,yt=[P[0]],[P[1]]
while True:
    iter2+=1
    P0=P
    P,err,loop=golden2D(P,u1,TOL)
    xt.append(P[0])
    yt.append(P[1])
    P,err,loop=golden2D(P,u2,TOL)
    xt.append(P[0])
    yt.append(P[1])
    
    u1=u2
    u2=P-P0
    amp=np.sqrt(sum(u2**2))
    if(amp<TOL):break
    P,err,loop=golden2D(P,u2/amp,TOL)
    xt.append(P[0])
    yt.append(P[1])

    
fig=plt.figure(figsize=(7,7))
plt.subplots_adjust(left=0.15,right=0.96,top=0.95,bottom=0.15,wspace=0.2,hspace=0.1)



x = np.linspace(-6, 5, 1000)
y = np.linspace(-5, 5, 1000)

X, Y = np.meshgrid(x, y)
E = ((X**2-10*Y)**2 + 2*Y**2 - 3*X)


dmax = np.max(E)
dmin = np.min(E)

levels = 6000*np.arange(20)/22  # (dmax-dmin)*np.arange(6000)/2900 +dmin
cmap = plt.cm.BuPu
con = plt.contourf(x, y, E, levels,cmap = cmap) 
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xt,yt,'b')
plt.scatter(xt,yt,color='yellow',edgecolors='k',alpha=1)
plt.show()
print("minimum point is ({:.3f},{:.3f})".format(xt[-1],yt[-1]))
print("iteration is {}".format(iter2))
```

```python
def func(x):
    return 100*(x[1] -x[0] **2) **2 + (1-x[0]) **2+100*(x[2] -x[1] **2) **2 + (1-x[2]) **2

def deriv(x): 
    f1=-400*x[0] *(x[1] -x[0] **2) -2*(1-x[0]) 
    f2=200*(x[1] -x[0] **2) -400*x[1] *(x[2] -x[1] **2) 
    f3=200*(x[2] -x[1] **2) -2*(1-x[2])
    return np.array([f1,f2,f3])

def golden2D(x,dr,TOL):
    R,err,loop=0.61803399,10.,-1
    dr = dr/np.sqrt(sum(dr**2))
    a= x-dr 
    b= x+dr
    while(err > TOL):
        loop += 1
        x1= b-R*(b-a)
        x2= a+R*(b-a)
        f1=func(x1)
        f2=func(x2)
        if(f2>f1): b=x2
        else: a=x1
        err=sum((a-b)**2)
        err=np.sqrt(err)
    xmin=a
    return xmin,err,loop

err = 1
TOL = 1.e-6
x = np.array([0., 2., 1.])
xt,yt,zt = [x[0]], [x[1]], [x[2]]

iteration = 0

while err>TOL:
    iteration += 1
    if iteration == 1:
        gradient_now = -deriv(x)
        dr_now = gradient_now
        xmin, err_iterate, loop_iterate = golden2D(x, dr_now, TOL)
        err = np.sqrt(sum((xmin-x)**2))
        x=xmin
        xt,yt,zt=np.append(xt,x[0]),np.append(yt,x[1] ),np.append(zt,x[2])
        gradient_prev = gradient_now
        dr_prev = dr_now
        
    else:
        gradient_now = -deriv(x)
        lam = sum(gradient_now**2)/sum(gradient_prev**2)
        dr_now = gradient_now + lam*dr_prev
        xmin, err_iterate, loop_iterate = golden2D(x, dr_now, TOL)
        err = np.sqrt(sum((xmin-x)**2))
        x=xmin
        xt,yt,zt=np. append(xt,x[0]),np. append(yt,x[1] ),np. append(zt,x[2])
        gradient_prev = gradient_now
        dr_prev = dr_now
    
print("minimum point :{} iteration : {}".format(xmin, iteration))
   
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(6,6))

ax = plt.axes(projection='3d')
ax.plot3D(xt,yt,zt, c='k', marker = 'o')
        
ax.set_xlim(-0.3, 1.2)
ax.set_ylim(-0.3, 2.2)
ax.set_zlim(-0.3, 1.5)  

        
```

```python
plt.figure(figsize=(8, 8)) 
ha_fft=np.fft.fft(v_ha) 
caii_fft=np.fft.fft(v_caii) 
Corr=np.fft.ifft(caii_fft*np.conj(ha_fft))
plt.plot(t,Corr.real, c='blue' ,marker='o' ,label=' FFT Correlate' ) 

Corr2=np.correlate(v_ha,v_caii,'full' ) 
lag=np.arange(-t[-1],t[-1]+0.01, 0.3)
plt.plot(lag,Corr2.real,c='red' ,marker='o',label='Numpy Correlate')
plt.xlim(0.,30.)
plt.ylim(-250.,350.) 
plt.xlabel('t [min]' )
plt.ylabel('Corr(t)' )
plt.legend()
```

FFT로 푸는 경우는 데이터의 범위가 $t=-\infty$ 에서 $t=\infty$ 까지 일 때 정확한 방법이다. 그러나 실제로는 데이터의 범위가 $t=0$ 에서 $t=30$ 까지이므로 정확한 결과 값이 아니다. 다른 방법으로 np.correlate를 사용할 수 있으며, 이 방법을 통해 얻은 결과는 조금 더 정확할 것이다. 하지만 두 방법으로 구한 결과 모두 경향성은 일치하는 것을 확인할 수 있다.
