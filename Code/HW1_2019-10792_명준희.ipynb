# 1. 두 개의 정수를 입력받아 몫과 나머지를 출력하는 프로그램

```python
a,b=(input('two integers: ').split('and')) #(정수 1) and (정수 2) 형식으로 입력해야만 한다. (hw1의 예시와 동일)
a=int(a)
b=int(b)
if (a>=b):
    c=int(a/b)
    d=a%b
else:
    c=int(b/a)
    d=b%a
print('quotient:',c)
print('remainder:',d)
```

# 2. n의 값을 입력받아 3n*3n 크기의 H 모양을 출력하는 프로그램

```python
n=int(input('n= '))
N=3*n
def topbot(n): #위와 아래의 n개의 행을 출력하는데 사용하는 함수
    for p in range(1,n+1):
        for i in range(1,N+1):
            if ((i<=n) or (i>2*n)):
                print("*",end="")
            if (i==N):
                print("")
            elif ((i>n) and (i<=2*n)):
                print(" ",end="")
def mid(n): #중간의 n개의 행을 출력하는데 사용하는 함수
    for q in range(1,n+1):
        for t in range(1,N+1):
            print('*',end="")
            if (t==N):
                print("")
topbot(n)
mid(n)
topbot(n)
```

# 3. 소수 판별 및 100,1000,10000,100000 이하의 소수의 개수를 알려주는 프로그램

```python
import numpy as np
import math as math
n=int(input("n="))
if (n<=1):
    print("2 이상의 자연수를 입력해 주십시오.")
def pdetector(n): #2 이상의 자연수 n을 입력받았을 때 그 수가 소수인지 판별하는 함수
    p=int(math.sqrt(n))
    d="소수"
    
    for i in range(2,p+1):
        if(n%i==0):
            d="합성수"
    return d
if(n>>1):
    print (n,"는(은)",pdetector(n),"입니다.")
def pnum(namx): #100, 1000, 100000 이하 소수의 개수를 출력하는 함수
    num=0
    for i in range(2,namx+1):
        if (pdetector(i)=="소수"):
            num=num+1
    return num
print ("100 이하 소수의 개수(2 제외) :",pnum(100)-1,"\n1000 이하 소수의 개수(2 제외) :",pnum(1000)-1,"\n10000 이하 소수의 개수(2 제외) :",pnum(10000)-1,"\n100000 이하 소수의 개수(2 제외) :",pnum(100000)-1)
```

# 4. 특정 데이터로 그래프를 만들고, 그 데이터의 4k+1번째 열만을 뽑아 저장하는 프로그램

```python
import numpy as np
import matplotlib.pyplot as plt
fn='./hw1_p4.dat'
x,a,b=np.loadtxt(fn,unpack=True,usecols=[0,1,2])
def fourth(a): #4k+1번째 행만 골라 새로운 배열을 만드는 함수
    a=np.array(a)
    la=len(a)
    A_4=np.zeros((2000,1))
    for i in range(1,la+1):
        if (i%5==1):
            p=int((i/5))
            A_4[p]=a[i-1]
    return A_4
X=fourth(x) #x,a,b에 대해 fourth 함수를 실행
A=fourth(a)
B=fourth(b)
ax=a*x #이 부분에서 그래프를 출력한다.
bx=b*x
abx=a*b*x
plt.plot(x,ax,'-',label='a*x')
plt.plot(x,bx,':',label='b*x')
plt.plot(x,abx,'--',label='a*b*x')
plt.xlabel('time')
plt.ylabel('value')
plt.legend(loc=2,numpoints=1)
plt.show()
file=open('hw1_p4_fourth.dat','w') #이 부분에서 fourth 함수로 변형한 데이터를 외부로 export한다.
for i in range(0,2000):
    xx=float(X[i])
    aa=float(A[i])
    bb=float(B[i])
    txt=("%6e %6e %6e\n"%(xx,aa,bb))
    file.write(txt)
file.close()
```
