```python
nmax = int(input("give the number nmax = "))

def primecheck(x):
    aliquot = 0
    for i in range(1, int(x**0.5)+1):#sqrt(x)까지만 나누어봐도 소수여부판단가능->속도 향상
        if x%i == 0:
            aliquot = aliquot + 1   
    if aliquot == 1:
        value = 1  # 소수, 자기 자신 빼고 나누어지는 것이 1 하나 밖에 없음
    else: 
        value = 0  # 소수 아님
    if x<2:#0, 1은 소수가 아님
        value = 0
    return value

count = 0
for i in range(3,nmax):
    if primecheck(i) == 1:
        count = count + 1
print("prime numbers between 2 and {} is {}".format(nmax, count))
print("단, 이 개수는 nmax와 2를 빼고 센 값이다.")

```

#### nmax = 100일 때 소수의 개수는 (2를 포함하면) 25개   
#### nmax = 1000일 때 소수의 개수는 (2를 포함하면)  168개 
#### nmax = 10000일 때 소수의 개수는 (2를 포함하면)  1229개
#### nmax = 100000일 때 소수의 개수는 (2를 포함하면)  9592개

```python
f2=open('hw1_p4_fourth.dat.txt', 'r')
import numpy as np
x,a,b=np.loadtxt(f2,unpack=True,usecols=[0,1,2])

import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(x, a*x, '-', label= 'a*x')
plt.plot(x, b*x, ':', label='b*x')
plt.plot(x, a*b*x, '--', label='a*b*x')
plt.legend(loc=2, numpoints=2, fontsize = 16)
plt.xlabel('time', fontsize = 20)
plt.ylabel('value', fontsize = 20)
plt.show()
# plot하는 코드 
```
