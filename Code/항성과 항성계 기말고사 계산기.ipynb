```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cal(a, b):
        C1 = 530 * a
        C2 = (1 - 1/0.2 * 1/b * a)
        return C1* C2


MrRatio = np.linspace(0, 0.3, 1000)

MstarRatio = 3

plt.plot(MrRatio, cal(MrRatio, MstarRatio))
plt.xlabel('$M_r/M_\odot$', fontsize = 15)
plt.ylabel('$L_r/L_\odot$', fontsize = 15)
plt.xlim(0, 0.3)
# plt.yscale('log')
plt.ylim(0, 100)
plt.xlim(0, 0.3)
```

### A.2.

```python
def X(r):
    if r<=0.3:

        x = 0.7*r
        t = 0.513*r*(1-(1/0.6)*r)
        son = x-t
        mother = r-t
        return son/mother

    else:
        return (0.7*r - 0.0775)/(r - 0.0775)


    

Mr = np.linspace(0, 3, 1000)

plt.ylim(0, 1)
for i in range(1000):
    if i == 1 :
        plt.scatter(Mr[i], X(Mr[i]), color = 'blue', s = 0.2, label = 'after $10^{8}yrs$')
    else:
        plt.scatter(Mr[i], X(Mr[i]), color = 'blue', s = 0.2)

        
plt.xlabel('$M_r/M_\odot$', fontsize = 15)
plt.ylabel('$X_{r}$', fontsize = 15)

plt.axhline(0.7, color = 'red', label = 'original')
plt.ylim(0, 0.75)
plt.xlim(0, 3)
plt.legend()
```
