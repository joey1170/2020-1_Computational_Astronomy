```python
#Chebyshev Polynomial
#Example

import numpy as np

def f(x):                        
    return np.sin(x)
order = 5
n = order + 1
# a, b 사이에서 f(x)를 추론해봐요 
a = 0
b = np.pi

x = np.linspace(a, b, 1000)
y = f(x)



############################# Chebyshev Polynomial ##########################
t0 = np.linspace(a, b, n)
for i in range(n):
    t0[i] = (np.pi/2)*(np.cos(np.pi*(2*i + 1)/(2*n+2)))



x_Cheb = t0
y_Cheb = f(x_Cheb)

p1 = np.polyfit(x_Cheb, y_Cheb, order)
poly1 = np.poly1d(p1)
y1 = poly1(x)


########################### even Spaced Nodes ###############################
x_even = np.linspace(a, b, n)
y_even = f(x_even)

p2 = np.polyfit(x_even, y_even, order)
poly2 = np.poly1d(p2)
y2 = poly2(x)

plt.plot(x, y1, label = 'Cheb')
plt.plot(x, f(x), label = 'original f(x)')
plt.plot(x, y2, label = 'Equal spaced')
plt.plot(x_Cheb, y_Cheb, 'ro')
plt.legend()
```

```python
#Chebyshev Polynomial
#Example

import numpy as np

def f(x):                        
    return 1/(1+x**2)
order = 30
n = order + 1
# a, b 사이에서 f(x)를 추론해봐요 
a = -5
b = 5

x = np.linspace(a, b, 1000)
y = f(x)



############################# Chebyshev Polynomial ##########################
t0 = np.linspace(a, b, n)
for i in range(n):
    t0[i] = 5*(np.cos(np.pi*(2*i + 1)/(2*n+2)))


#linear change of variables
x_Cheb =t0
y_Cheb = f(x_Cheb)

#interpolation
p1 = np.polyfit(x_Cheb, y_Cheb, order)
poly1 = np.poly1d(p1)
y1 = poly1(x)


########################### Even Spaced Nodes ###############################
x_even = np.linspace(a, b, n)
y_even = f(x_even)

p2 = np.polyfit(x_even, y_even, order)
poly2 = np.poly1d(p2)
y2 = poly2(x)

plt.plot(x, y1, label = 'Cheb')

plt.plot(x, f(x), label = 'original f(x), order= {}'.format(order))
plt.plot(x_Cheb, y_Cheb, 'ro')
plt.legend()
```

```python
from scipy import interpolate

x = [0, 1, 2, 2.5, 3, 3.5, 4]
y = [2.5, 0.5, 0.5, 1.5, 1.5, 1.125, 0]


result1 = interpolate.CubicSpline(x, y, bc_type = 'not-a-knot')
result2 = interpolate.CubicSpline(x, y, bc_type = 'clamped')
result3 = interpolate.CubicSpline(x, y, bc_type = 'natural')
xs= np.linspace(0, 4, 1000)
plt.plot(xs, result1(xs))
plt.plot(xs, result2(xs))
plt.plot(xs, result3(xs))
plt.plot(x, y, 'ro')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

#raw data
x, y = np.mgrid[-1:1:20j, -1:1:20j]
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
xx, yy = np.meshgrid(x, y)
z = (xx + yy)*np.exp(-6.0*(xx*xx + yy*yy))
plt.pcolor(xx, yy, z)
plt.colorbar()
plt.title('raw data')
plt.show()


#Cubic Spline
xnew = np.linspace(-1, 1, 200)
ynew = np.linspace(-1, 1, 200)
f = interpolate.interp2d(x, y, z, kind = 'cubic')
znew = f(xnew, ynew)

plt.pcolor(xnew, ynew, znew)
plt.colorbar()
plt.title("Interpolated data-cubic spline")
plt.show()

#B Spline
tck = interpolate.bisplrep(xx, yy, z, s = 0)
znew = interpolate.bisplev(xnew, ynew, tck)

plt.pcolor(xnew, ynew, znew)
plt.colorbar()
plt.title("Interpolated Data-B spline")
plt.show()
```
